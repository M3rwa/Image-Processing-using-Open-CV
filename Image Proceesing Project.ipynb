```python
#Image Enhancement(Image Denoising): The process of manipulating an image so that the result is more suitable than the original for a specific application.

import cv2

def denoise_image(image_path):
    # Load the image
    image = cv2.imread(image_path)
    
    # Apply denoising techniques
    denoised_image = cv2.fastNlMeansDenoisingColored(image, None, 10, 10, 7, 21)
    
    # Display the original and denoised images
    cv2.imshow("Original Image", image)
    cv2.imshow("Denoised Image", denoised_image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

# Path to the image file
image_path = "C:\\Users\\MARWA\\Desktop\\SEM 6\\6.Medical Information Visualization\\Project\\Data\\noisy medical image.png"
#image_path = "C:\\Users\\MARWA\\Desktop\\SEM 6\\6.Medical Information Visualization\\Project\\Data\\Unsharp.jpg"

# Call the denoise_image function
denoise_image(image_path)
```

```python
#Image Sharpening :Sharpening enhances the definition of edges in an image.

import cv2
import numpy as np

def sharpen_image(image_path):
    # Load the image
    image = cv2.imread(image_path)
    
    # Create the sharpening kernel
    kernel = np.array([[0, -1, 0],
                       [-1, 5, -1],
                       [0, -1, 0]])
    
    # Apply the kernel to the image
    sharpened_image = cv2.filter2D(image, -1, kernel)
    
    # Display the original and sharpened images
    cv2.imshow("Original Image", image)
    cv2.imshow("Sharpened Image", sharpened_image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

# Path to the image file
image_path = "C:\\Users\\MARWA\\Desktop\\SEM 6\\6.Medical Information Visualization\\Project\\Data\\Unsharp.jpg"

# Call the sharpen_image function
sharpen_image(image_path)
```

```python
# morphological processing : Tools for extracting image components that are useful in the representation and description of shapes.

import cv2

# Read the image
image_path = "C:\\Users\\MARWA\\Desktop\\SEM 6\\6.Medical Information Visualization\\Project\\Data\\enhancement.png"
image = cv2.imread(image_path)

# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply contrast enhancement to the image
enhanced_image = cv2.equalizeHist(gray_image)

# Display the original and enhanced images
cv2.imshow("Original Image", gray_image)
cv2.imshow("Enhanced Image", enhanced_image)

# Wait for a key press and then close the windows
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Image segmentation : Partition an image into its constituent parts or objects.The more accurate the segmentation, the more likely recognition is to succeed.
 
import cv2

# Read the image
image_path = "C:\\Users\\MARWA\\Desktop\\SEM 6\\6.Medical Information Visualization\\Project\\Data\\segmenting .png"

try:
    image = cv2.imread(image_path)

    if image is None:
        raise FileNotFoundError(f"Failed to load image: {image_path}")

    # Convert the image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply image segmentation techniques (e.g., thresholding)
    # Example: Thresholding with a fixed threshold value
    _, thresholded_image = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)

    # Display the original and segmented images
    cv2.imshow("Original Image", image)
    cv2.imshow("Segmented Image", thresholded_image)

    # Wait for a key press and then close the windows
    cv2.waitKey(0)
    cv2.destroyAllWindows()

except Exception as e:
    print(f"An error occurred: {e}")
```
